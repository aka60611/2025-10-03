```python
import random

def output(jan):
    for i in range(len(jan)):
        if i == 0:
            p = "あなた"
        else:
            p = "CPU"
        if jan[i] == 0:
            r = "グー"
        elif jan[i] == 1:
            r = "チョキ"
        else:
            r = "パー"
        print(p + "->" + r)

def jannken(jan):
    if jan[0] == jan[1]:
        result = 0
        print("あいこ")
    elif jan[0] + 1 == jan[1] or jan[0] - 2 == jan[1]:
        result = 1
        print("勝ち")
    else:
        result = 2
        print("負け")
    return result


print("---じゃんけんゲーム---")
while True:
    s=0
    while s == 0:
        you = input("半角数字で 0,1,2 のどれかを入力してください (0=グー 1=チョキ 2=パー)")
        if you == "0" or you == "1" or you == "2":
            s += 1
        else:
            print("0,1,2以外の文字が入力さえれています")

    cpu = random.randrange(3)
    jan = [int(you), cpu]
    output(jan)
    result = jannken(jan)
    if result != 0:
        break

```
